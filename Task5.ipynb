```python
import pandas as pd
df = pd.read_csv("train.csv")
df.info()
print()
```

```python
print(df.describe())
print()
print(df.isnull().sum())
print()
print("No. of Survivals:")
print(df['Survived'].value_counts())
print()
print("No. of passangers of particular gender:")
print(df['Sex'].value_counts())
print()
print("No. of passangers in different class :")
print(df['Pclass'].value_counts())
```

```python
import seaborn as sb
import matplotlib.pyplot as mpt
%matplotlib inline
sb.histplot(df['Age'].dropna(),bins = 30, kde = True)
mpt.title("Age Distribution")
mpt.xlabel("Age")
mpt.ylabel("Count")
mpt.show()
```

```python
sb.countplot(x= 'Pclass',data = df)
mpt.title("Passanger Class distribution")
mpt.xlabel("Pclass")
mpt.ylabel("Count")
mpt.show()
```

```python
sb.countplot(x = 'Sex', data = df)
mpt.title("Gender Count")
mpt.xlabel("Sex")
mpt.ylabel("Count")
mpt.show()
```

```python
sb.countplot(x = 'Survived' , data = df)
mpt.title("No. of Survivors")
mpt.xlabel("Survived (0 = No ,1= Yes)")
mpt.ylabel("Count")
mpt.show()
print()
```

```python
mpt.figure(figsize=(10,7))
sb.countplot(data =df , x = 'Pclass' , hue = 'Survived')
mpt.title("Survival by Class of Passanger ")
mpt.xlabel("Pclass")
mpt.ylabel("Count")
mpt.tight_layout()
mpt.subplots_adjust(bottom = 0.5)
mpt.show()
```

```python
corr= df[['Survived', 'Pclass','Age' ,'SibSp','Parch','Fare']].corr()
mpt.figure(figsize = (10,7))
sb.heatmap(corr , annot = True , cmap = 'coolwarm', linewidths = 0.7)
mpt.title("Correlation Matrix")
mpt.tight_layout()
mpt.show()
```

```python
numerical_features = ['Survived', 'Pclass','Age' , 'SibSp' , 'Parch','Fare']
sb.pairplot(df[numerical_features] , hue = 'Survived' , palette = 'coolwarm')
mpt.suptitle('Pair_plot of Selected Features' , y = 1.02)
mpt.show()
```
